```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
```

```python
houses_ds = sklearn.datasets.fetch_california_housing(as_frame=True)['frame']
houses_ds.head(5)
```

```python
houses_ds.describe()
```

```python
houses_ds.hist(figsize=(20,15), bins=50)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(houses_ds, test_size=0.2, random_state=21)
```

```python
# adding income_category attribute for stratified sampling

import numpy as np
houses_ds['income_category'] = pd.cut(houses_ds['MedInc'],
                                     bins = [0,1.5,3,4.5,6,7.5,np.inf],
                                     labels = [1,2,3,4,5,6])
houses_ds['income_category'].hist()
plt.show()
```

```python
# stratified sampling of income_category attribute

from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=21)
for train_index, test_index in split.split(houses_ds, houses_ds['income_category']):
    stratified_train_set = houses_ds.loc[train_index]
    stratified_test_set = houses_ds.loc[test_index]
```

```python
# comparison of income_category proportions in the test_set and the original dataset

print(stratified_test_set['income_category'].value_counts() / len(stratified_test_set))
print(houses_ds['income_category'].value_counts() / len(houses_ds))
```

```python
fig = stratified_train_set.plot(kind='scatter', x='Longitude', y='Latitude', alpha = 0.4,
                          s=stratified_train_set['Population']/100, label='population',
                          c = 'MedHouseVal',figsize=(10,7), cmap = plt.get_cmap("plasma"),
                          colorbar=True
                         )
plt.legend()
```

```python
# correlations between attributes and median price

corr_matrix = stratified_train_set.corr()
corr_matrix['MedHouseVal'].sort_values(ascending=False)
```

```python
# correlation matrix of 4 most promising attributes

pd.plotting.scatter_matrix(stratified_train_set[['MedHouseVal','MedInc','AveRooms','HouseAge']],
                           figsize=(12,8))
plt.show()
```

```python
# median income vs median house value
stratified_train_set.plot(kind='scatter', x='MedHouseVal', y='MedInc', alpha=0.1)
plt.show()
```

```python
# separating attributes and labels
train_labels = stratified_train_set['MedHouseVal']
test_labels = stratified_test_set['MedHouseVal']
stratified_train_set.drop('MedHouseVal', axis=1, inplace=True)
stratified_test_set.drop('MedHouseVal', axis=1, inplace=True)
```

```python
# scaling the data
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

transform_pipeline = Pipeline([
    ('std_scaler', StandardScaler())
])
train_transformed = transform_pipeline.fit_transform(stratified_train_set)
test_transformed = transform_pipeline.fit_transform(stratified_test_set)
```

```python
from sklearn.metrics import mean_squared_error

# training model
def fit_model(model):
    model.fit(train_transformed, train_labels)

    sample_data = train_transformed[100:105]
    sample_labels = train_labels[100:105]

    print(f"Predicted: {model.predict(sample_data)}")
    print(f"Labels: {sample_labels.to_numpy()}")
    
# measuring the error
def measure_train_error(model):
    predictions = model.predict(train_transformed)
    rmse = np.sqrt(
        mean_squared_error(train_labels, predictions)
    )
    print(f"RMSE: {rmse:.3f}")
    print(f"(RMSE / Mean_house_value) * 100%: {rmse / train_labels.mean() * 100:.2f}%")
```

```python
from sklearn.linear_model import LinearRegression

# linear regression
linear_reg = LinearRegression()

print('Linear regression')
fit_model(linear_reg)
measure_train_error(linear_reg)
```

```python
# decision tree regression
from sklearn.tree import DecisionTreeRegressor

dtree_reg = DecisionTreeRegressor()

print('Decision tree')
fit_model(dtree_reg)
measure_train_error(dtree_reg)
```

```python
# cross validation
from sklearn.model_selection import cross_val_score

def cross_validation(regressor):
    rmses = -cross_val_score(
        regressor, train_transformed, train_labels,
        scoring = 'neg_mean_squared_error', cv=5
    )
    rmses_scores = np.sqrt(rmses)

    print('Cross validation scores:', rmses_scores)
    print(f'Mean: {rmses_scores.mean():.2f}')
    print(f"(Mean RMSE / Mean_house_value) * 100%: {rmses_scores.mean() / train_labels.mean() * 100:.2f}%")
```

```python
# performing cross validation for a decision tree 
# to compare with the result of the linear regression

print('Decision tree')
cross_validation(dtree_reg)

print('\n', 'Linear regression')
cross_validation(linear_reg)
```

```python
# random forest regression
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor()

print('Random Forest')
fit_model(forest_reg)
measure_train_error(forest_reg)
cross_validation(forest_reg)
```

```python
# grid search for decision tree
from sklearn.model_selection import GridSearchCV

param_grid = [
    {'max_features': [2, 4, 6], 'max_depth': [2, 3, 4]}
]

grid_search = GridSearchCV(
    DecisionTreeRegressor(),
    param_grid, cv = 5,
    scoring = 'neg_mean_squared_error',
    return_train_score=True
)
grid_search.fit(train_transformed, train_labels)

results = grid_search.cv_results_
print(f'best score and params: {np.sqrt(-grid_search.best_score_):.3f}, {grid_search.best_params_}')
for mean_score, params in zip(results['mean_test_score'], results['params']):
    print(f'score {np.sqrt(-mean_score):.3f}, params {params}')
```

```python
# feature importance
feature_importance = grid_search.best_estimator_.feature_importances_
feature_names = stratified_train_set.columns.values

for i in sorted(zip(feature_importance, feature_names), reverse=True):
    print(i)
```

```python
# measuring error on the test set

def measure_test_error(model, name):
    predictions = model.predict(test_transformed)
    rmse = np.sqrt(
        mean_squared_error(test_labels, predictions)
    )
    print(f'{name}')
    print(f"Test_RMSE: {rmse:.3f}")
    print(f"(Test_RMSE / Mean_house_value) * 100%: {rmse / test_labels.mean() * 100:.2f}%")
    print()
```

```python
measure_test_error(linear_reg, 'Linear regression')
measure_test_error(dtree_reg, 'Decision tree')
measure_test_error(forest_reg, 'Random forest')
```
