```python
class LinkedList:
    def __init__(self, arr):
        self.head = None
        self.tail = None
        
        prev = None
        for i in arr:
            node = Node(
                key_=i,
                next_=None,
                prev_=prev
            )
            if prev:
                prev.next = node
            prev = node
            
            if not self.head:
                self.head = node
        self.tail = prev
            
    def traverse_forward(self):
        elt = self.head
        while True:
            if elt:
                print(elt.key)
                elt = elt.next
            else:
                break
                
    def traverse_backward(self):
        elt = self.tail
        while True:
            if elt:
                print(elt.key)
                elt = elt.prev
            else:
                break
    
    def search(self, key):
        elt = self.head
        position = None
        curr_position = 0
        result_node = None
        
        while True:
            if elt:
                if elt.key == key:
                    position = curr_position
                    result_node = elt
                    break
                else:
                    curr_position += 1
                elt = elt.next
            else:
                break
        return position, result_node
    
    def insert(self, key, position):
        elt = self.head
        i = position
        while True:
            if elt:
                elt = elt.next
                i -= 1
            else:
                raise ValueError("Inserting to position > list length")
            if not i:
                break
        if elt:
            new_node = Node(
                key_ = key,
                next_ = elt,
                prev_ = elt.prev
            )
            if elt.prev:
                elt.prev.next = new_node
            elt.prev = new_node 
        else:
            new_node = Node(
                key_ = key,
                next_ = None,
                prev_ = None
            )
        if not position:
            self.head = new_node
        
    
    def delete(self, key):
        pos, elt = self.search(key)
        if elt.prev:
            elt.prev.next = elt.next
        if elt.next:
            elt.next.prev = elt.prev
        if elt == self.head:
            self.head = elt.next
        if elt == self.tail:
            self.tail = elt.prev
        del(elt)
        
        
class Node:
    def __init__(self, key_, next_, prev_):
        self.key = key_
        self.next = next_
        self.prev = prev_
```

```python
llist = LinkedList(
    [2,3,1,6,5,4]
)
llist.traverse_forward()
print()
llist.traverse_backward()
print()
print(llist.search(6))

llist.insert(42, 5)
llist.traverse_forward()

print()
llist.delete(2)
llist.traverse_forward()
```
