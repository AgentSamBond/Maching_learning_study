```python
import igraph
import matplotlib.pyplot as plt
```

```python
def dfs(G, v, parent, path_length, path_sequence):
    print(path_sequence)
    if len(path_sequence) == path_length:
        return path_sequence
    
    n_paths = []
    
    nodes_to_traverse = v.neighbors().copy()
    if parent in nodes_to_traverse:
        nodes_to_traverse.remove(parent)

    for i in nodes_to_traverse:
        curr_path = path_sequence.copy()
        curr_path.append(v)
        paths = dfs(G, i, v, path_length, curr_path)
        if paths:
            n_paths += paths
        
    return n_paths
```

```python
# fig, ax = plt.subplots()
igraph.config['plotting.backend'] = 'matplotlib'

for g in [
    igraph.Graph(n=4, edges=[[0, 1], [1, 2], [2, 3], [1, 3]]),
    igraph.Graph(n=4, edges=[[0, 1], [1, 2], [2, 3], [3, 0], [1, 3]])
    
]:
    igraph.plot(g)
    simple_paths = dfs(g, g.vs[0], None, 3, [])
    plt.title(f'simple paths: {len(simple_paths)}')
    plt.show()
    plt.clf()
```
